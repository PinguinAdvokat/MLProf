```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import sqlite3 as sq

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

# 1.1 Загрузка данных

```python
data = pd.read_csv("datasets/data.csv")
data = data.drop(['Unnamed: 32'], axis=1)
```

```python
data
```

### Кол-во Nan значений

```python
data.isnull().sum()
```

### сохраняем датасет в базу данных

```python
try:
    con = sq.connect("data.db")
except Exception as e:
    print(e)
```

```python
try:
    data.to_sql("raw_data", con=con, if_exists="append", index=False)
except Exception as e:
    print(e)
```

# 1.2 Предобработка данных и выделение значимых атрибутов

заменяем категориальные значения на числовые

```python
data['diagnosis'] = data['diagnosis'].map({'M': 0, 'B': 1})
```

Кластеризация с использованием KMEans и PCA

```python
pca = PCA(n_components=2)
```

```python
data_2d = pca.fit_transform(data)
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data_2d)
clusters = kmeans.predict(data_2d)
```

```python
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=clusters)
plt.show()
```

### Добавляем кластеры в датасет

```python
clusters = pd.Series(clusters)
clusters.name = 'clusters'
data = pd.concat([data, clusters], axis=1)
```

```python
data
```

# 1.5 Формирование итогового набора данных

### Обучаем простую модель дерева выбора

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('diagnosis', axis=1), data['diagnosis'], test_size=0.33, random_state=42)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
tree.score(X_test, y_test)
```

### Значимые признаки на основе обученной модели

```python
plt.bar(list(X_train.columns), tree.feature_importances_)
plt.xticks(rotation=90)
plt.show()
```

# 1.4 Определение численных характеристик наборов данных

Для всех diagnosis найти среднее значение texture_se

```python
data.groupby('diagnosis')['texture_se'].mean()
```

Для каждой записи найти сумму area_se + perimeter_se + среднее значение по всем texture_se

```python
mean = data['texture_se'].mean()
pd.Series(data['area_se'] + data['perimeter_se'] + mean)
```

Для каждой записи вывести произведение texture_se на area_se если значение diagnosis = B. Иначе выводить сумму texture_se + area_se

```python
data.apply(lambda data: data['texture_se'] * data['area_se'] if data['diagnosis'] == 1 else data['texture_se'] + data['area_se'], axis=1)
```
